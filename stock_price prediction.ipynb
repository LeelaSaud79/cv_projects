```python
df['Mean avg 30 day']=df['Close'].rolling(window=30).mean()
df['Standard Deviation']=df['Close'].rolling(window=20).std()
```

```python
df['Upper Band']=df['Mean avg 30 day']+(df['Standard Deviation']*2)
df['Lower Band']=df['Mean avg 30 day']=(df['Standard Deviation']*2)
df.tail()
```

```python
df[['Adj Close', 'Mean avg 30 day', 'Upper Band', 'Lower Band']].plot(figsize=(12,5))
plt.title('Bollinger Band plot for overall period ', fontsize=18)
```

Train and Testing Data

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
```

```python
df.columns
```

```python
x=df[['Open', 'High', 'Low', 'Close', 'Volume']]
y=df['Adj Close']
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(x, y, test_size=0.2, random_state=32)
xtrain.head()
```

Scaling of data

```python
scale = StandardScaler()
xtrain = scale.fit_transform(xtrain)
xtest = scale.fit_transform(xtest)
xtrain
```

```python
xtest
```

Applying the model

```python
model = RandomForestRegressor(n_estimators=100, random_state = 42, max_depth=10)
model.fit(xtrain, ytrain)
```

```python
predict=model.predict(xtest)
predict
```

Predictions:

```python
predictions = pd.DataFrame({"Predictions": predict}, index=pd.date_range(start=df.index[-1], periods=len(predict), freq="D"))
predictions.to_csv("Predicted-price.csv")
```

```python
# collecting future days from predicated values
oneyear_df= pd.DataFrame(predictions[:240])
oneyear_df.to_csv("oneyear_prediction.csv")
onemonth_df= pd.DataFrame(predictions[:30])
onemonth_df.to_csv("onemonth_prediction.csv")
oneweek_df= pd.DataFrame(predictions[:7])
oneweek_df.to_csv("oneweek_prediction.csv")
oneday_df= pd.DataFrame(predictions[:1])
oneday_df.to_csv("day_prediction.csv")
```
