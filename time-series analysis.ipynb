```python
df['avg_temp']=(df['JAN']+df['FEB']+df['MAR']+df['APR']+df['MAY']+df['JUN']+df['JUL']+df['AUG']+df['SEP']+df['OCT']+
               df['NOV']+df['DEC'])/12
```

```python
plt.style.use('seaborn-whitegrid')
plt.title('Average title in India')
plt.plot(df.YEAR, df.avg_temp,'g.-', label='Average Temperature')
plt.xlabel('Year')
plt.ylabel('Temperature in celsius')
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')
plt.title(' temperature in India from 1901 to 2017')

plt.plot(df.YEAR, df.JAN, label = 'January')
plt.plot(df.YEAR, df.FEB, label = 'February')
plt.plot(df.YEAR, df.MAR, label = ' March')
plt.plot(df.YEAR, df.APR, label = 'April')
plt.plot(df.YEAR, df.MAY, label = 'May')
plt.plot(df.YEAR, df.JUN, label = 'June')
plt.plot(df.YEAR, df.JUL, label = 'July')
plt.plot(df.YEAR, df.AUG, label = 'August')
plt.plot(df.YEAR, df.SEP, label = 'September')
plt.plot(df.YEAR, df.OCT, label = 'October')
plt.plot(df.YEAR, df.NOV, label = 'November')
plt.plot(df.YEAR, df.DEC, label = 'December')


plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.xticks(df.YEAR[::29])

plt.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)
```
